## Inlab: Feature Selection based on Correlation analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
df
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df.corr()
```

```python
import seaborn as sb
dataplot = sb.heatmap(df.corr(), cmap="YlGnBu")
plt.show()
```

```python
#Keep only the features which are having correlation with price > abs (0.5) 
uncorrelated_features = set()
correlation_matrix = df.corr()
correlation_matrix
```

```python
i = 'SalePrice'
for j in list(correlation_matrix.columns):
    if abs(correlation_matrix.loc[i, j]) <= 0.5:
        colname = j
        uncorrelated_features.add(colname)
```

```python
uncorrelated_features
```

```python
len(uncorrelated_features)
```

```python
df.drop(labels=uncorrelated_features, axis=1, inplace=True)
```

## Post Lab: PCA (Principle Component Analysis)

```python
#Post Lab
import pandas as pd
import numpy as np
```

```python
post = pd.read_csv('Iris.csv')
post
```

```python
#Divide the dataset into featues and target
X = post.iloc[:,0:-1].values
y = post.iloc[:,-1].values
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# performing preprocessing part
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
 
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Applying PCA function on training
# and testing set of X component
from sklearn.decomposition import PCA
 
pca = PCA(n_components = 2)
 
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
 
explained_variance = pca.explained_variance_ratio_
```
