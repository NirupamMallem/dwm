```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

<h2> Inlab 1 </h2>

```python
#Load the dataset. 
data = pd.read_csv('Social_Network_Ads.csv')
data
```

```python
data.isna().sum()
```

```python
data.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['Gender'] = le.fit_transform(data['Gender'])
data
```

```python
# Find out dependent and independent variables. 
X = np.array(data.iloc[:,1:-1])
y = np.array(data.iloc[:,-1]).reshape(-1,1)
```

```python
# Split dataset into train and test
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
#Fit the model in the Decision Tree classifier. 
from sklearn.tree import DecisionTreeClassifier
Predtree = DecisionTreeClassifier(criterion='entropy',max_depth=3)
Predtree.fit(X_train,y_train)
```

```python
#Make Predictions
y_pred = Predtree.predict(X_test)
y_pred
```

```python
#Check accuracy
from sklearn.metrics import *
accuracy_score(y_test,y_pred)
```

```python
#Load and Explore the Dataset. 
df = pd.read_csv('adult.csv')
df
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
#Perform required pre-processing
df.columns = ['age', 'workclass', 'fnlwgt', 'education', 'never_married', 'marital_status', 'occupation', 'relationship',
             'race', 'sex', 'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'income']
df.head()
```

```python
df.income = pd.get_dummies(df.income)[' >50K']
```

```python
df['income']
```

```python
categorical_names = []
for feature in df.columns: 
    if df[feature].dtype == object: 
        categorical_names.append(feature)
categorical_names
```

```python
df[categorical_names].head() 
```

```python
for feature in df[categorical_names].columns:
    print('FEATURE NAME:', feature)
    print(df[feature].value_counts())
```

```python
for feature in df.columns:
    df[feature].replace(' ?', np.nan, inplace=True)
```

```python
df[categorical_names].isnull().any() 
```

```python
numerical_features = [var for var in df.columns if df[var].dtype!='O']

df[numerical_features].head() 
```

```python
na_colls = df.isnull().any().loc[df.isnull().any().values == True].index
na_colls
```

```python
for i in na_colls:
    df[i].fillna(df[i].mode()[0], inplace=True)

df.isnull().any()
```

```python
import category_encoders as ce
categorical_names_withoutone = categorical_names
categorical_names_withoutone.remove('native_country')
encoder = ce.OneHotEncoder(cols=categorical_names_withoutone)

data_encoded = encoder.fit_transform(df)

data_encoded.head()
```

```python
print('native_country has got', df.native_country.value_counts().shape[0], 'features.')
```

```python
mean_encoded_nativeCont = data_encoded.groupby(['native_country'])['income'].mean().to_dict() 
data_encoded.native_country = data_encoded.native_country.map(mean_encoded_nativeCont)
```

```python
data_encoded.native_country
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
target = data_encoded.income 
features = data_encoded.drop('income', axis=1) 

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.33)

gnb = GaussianNB() 
gnb.fit(X_train, y_train)
```

```python
prediction = gnb.predict(X_test)

prediction
```

```python
correct = (y_test == prediction).sum() 
print('classified correctly', correct) 
wrong = X_test.shape[0] - correct 
print('classified incorretly', wrong)
```

```python
print('The Accuracy is', correct / X_test.shape[0])
```
