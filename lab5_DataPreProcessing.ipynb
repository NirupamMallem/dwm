```python
#Lets impute all objects with their mode
from sklearn.impute import SimpleImputer
Objectimputer = SimpleImputer(missing_values = np.nan,
                        strategy ='most_frequent')
```

```python
Objectimputer = Objectimputer.fit(df[['director_name','plot_keywords','content_rating']])
df[['director_name','plot_keywords','content_rating']]= Objectimputer.transform(df[['director_name','plot_keywords','content_rating']])
```

```python
df.isna().sum()
```

```python
#Lets also impute title year, director facebook likes with mode
Objectimputer = Objectimputer.fit(df[['director_facebook_likes','title_year']])
df[['director_facebook_likes','title_year']]= Objectimputer.transform(df[['director_facebook_likes','title_year']])
```

```python
df.isna().sum()
```

```python
#Lets impute every other column with missing values more than 100 with mean
imputer = SimpleImputer(missing_values = np.nan,
                        strategy ='mean')
```

```python
imputer = imputer.fit(df[['gross','budget','aspect_ratio']])
df[['gross','budget','aspect_ratio']] = imputer.transform(df[['gross','budget','aspect_ratio']])
```

```python
df.isna().sum()
```

```python
#Remove the rows with NaN values in remaining columns
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
#Label Encode the columns ‘language’, ‘country’, ‘content rating’
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()

df['language'] = label_encoder.fit_transform(df['language'])
```

```python
df['country'] = label_encoder.fit_transform(df['country'])
df['content_rating'] = label_encoder.fit_transform(df['content_rating'])
```

```python
df['language'].unique()
```

```python
df.dtypes
```

```python
#OneHotEncode the column ‘genres’ 
df = pd.get_dummies(df, columns = ['genres'])
df
```

```python
'''
Find the missing values of hours and distance and replace with the median and find
the correlation and visualize it. And find the correlation by normalizing the data using
min-max normalization technique. And find the difference in both correlations.
'''
post2 = post.copy()
```

```python
post2.isna().sum()
```

```python
Medianimputer = SimpleImputer(missing_values = np.nan,
                        strategy ='median')
```

```python
Medianimputer = Medianimputer.fit(post2[['hour']])
post2[['hour']] = Medianimputer.transform(post2[['hour']])
post2.isna().sum()
```

```python
post2[['distance','hour']].corr()
```

```python
import seaborn as sb
plot = sb.heatmap(data=post2[['distance','hour']].corr())
plt.show()
```

```python
post2['distance'] = (post2['distance'] - post2['distance'].min()) / (post2['distance'].max() - post2['distance'].min())
post2['hour'] = (post2['hour'] - post2['hour'].min()) / (post2['hour'].max() - post2['hour'].min())
```

```python
post2[['distance','hour']].corr()
```

Again there is no difference in correlation value
